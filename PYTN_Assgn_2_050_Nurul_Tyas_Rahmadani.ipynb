Now we can analyze Confidence Intervals below.

Function below are graded function. (1 Points)

```python
# Write your function below

# Graded-Funtion Begin
sale_price = df.groupby("BOROUGH").agg({"SALE PRICE": [np.mean, np.std, np.size]})
print(sale_price)
print('-'*75)
print('                     CONFIDENCE INTERVAL UNTUK SALE PRICE')
print('-'*75)

ci95_hi = []
ci95_lo = []
z = 1.96 %ambil selang kepercayaan 95%

for i in sale_price.index:
    m,s,n = sale_price.loc[i]
    ci95_lo.append (m - z*s/np.sqrt(n))
    ci95_hi.append (m + z*s/np.sqrt(n))

sale_price['ci95_lo'] = ci95_lo    
sale_price['ci95_hi'] = ci95_hi

print(sale_price)


# Graded-Funtion End
```

Diperoleh confidence interval dengan CI 95% untuk masing2 sale price borough 1-5 pada tabel di atas

Make your Hypothesis Testing below

Function below are graded function. (1 Points)

Akan diuji hipotesis antara sale price borough 1 dan borough2.
Null Hipotesis mu1 = mu2
Alternative Hipotesis mu 1 != mu 2

```python
# Write your function below
import statsmodels.api as sm

# Graded-Funtion Begin
borough1 = df[df["BOROUGH"]==1]
#print(borough1)
borough2 = df[df["BOROUGH"]==2]
#print(borough2)

n1 = len(borough1)
print('n1: ',n1)
mu1 = borough1["SALE PRICE"].mean()
print('mu1: ',mu1)
sd1 = borough1["SALE PRICE"].std()
print('sd1: ',sd1)
print('-'*50)

n2 = len(borough2)
print('n2: ',n2)
mu2 = borough2["SALE PRICE"].mean()
print('mu2: ',mu2)
sd2 = borough2["SALE PRICE"].std()
print('sd2: ',sd2)
print('-'*50)

sm.stats.ztest(borough1["SALE PRICE"].dropna(), borough2["SALE PRICE"].dropna(), alternative='two-sided')


# Graded-Funtion End
```
